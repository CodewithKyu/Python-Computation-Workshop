# Matplotlib

* a Python library, specially designed for the development of graphs, chats etc. to provide interactive data visualization
* It is inspired from the MATALB and reproduces many of its features

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10)
y = x*x
```

```python
plt.plot(x,y)
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x, y, marker='o', markersize=2, linestyle='None', linewidth=2, color='red', label='Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='Sin(x)', color='blue')
plt.plot(x, y2, label='Cos(x)', color='red', linestyle='--')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
## Histogram

y = np.random.randn(100,100)
plt.hist(y)
plt.show()
```

```python
# Histogram using bin
y = np.random.randn(100000)
plt.hist(y,100, color='red')
plt.show()
```

```python
# Bar chart

x = [1,2,3]
y = [1,4,9]
plt.bar(x,y, color='green')
plt.show()
```

```python
# Pie chart
plt.figure()
x = [40,20,5, 30]
labels = ['Cycles', 'Bikes', 'Cars', 'Buses']
plt.pie(x,labels=labels)
plt.show()
```
